```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'same') / w
```

```python
def exp(x, A, tau, c):
    return A*np.exp(-x/tau) + c
```

```python
data = pd.read_excel('LED_power.xlsx').iloc[170:]
data.reset_index(inplace=True)

step = 0.06467
data['t'] =  np.linspace(0, step*len(data['t']), len(data['t']), endpoint=False)
data['index'] = np.array(data['index']-170)

t = np.array(data['t'])
```

# tON tOFF detection

```python
# %matplotlib notebook
%matplotlib inline
plt.plot(data['index'], data['bg'])
plt.xlim(1000, 1100)
plt.show()
```

```python
on = []
off=[]

flag = 0
for i in range(len(data.t)):
    if i<=1410:
        if ((flag == 0) and (data['bg'][i]>=3)):
            flag = 1
            on.append(i-1)
        if ((flag == 1) and (data['bg'][i]<3)):
            flag = 0
            off.append(i)
    if i>1410:
        if ((flag == 0) and (data['bg'][i]>=1.5)):
            flag = 1
            on.append(i-1)
        if ((flag == 1) and (data['bg'][i]<1.5)):
            flag = 0
            off.append(i)
    
on.append(len(data.t)-1)
off = list(np.array(off)-1)
```

```python
plt.figure(figsize=(40, 20), dpi=1000)
fig, axs = plt.subplots(1,1)
#wt
# axs.plot(data.t, data['bg'])
axs.plot(data['index'], data['bg'], color='black', label='bg')
axs.plot(data['index'], data['Intensity Region 2'], color='blue', label='wt')

for i in range(len(off)):
    axs.plot(data['index'][off[i]:on[i+1]], data['Intensity Region 2'][off[i]:on[i+1]],color='red')#, label='mut-FIT')
#     axs.scatter(on[:-1], downs[0]*5)

           
# #mut
# axs[1].plot(data.t, data['bg-mut']*5, color='grey', label='bg-mut')
# axs[1].plot(data.t, data['D220N-1'], color='green', label='D220N')
# for i in range(len(off_D)):
#     axs[1].plot(data.t[off_D[i]:on_D[i+1]], data['D220N-1'][off_D[i]:on_D[i+1]],color='orange')#, label='mut-FIT')
    
#sns.catplot(data=data, x= data.iloc[:, 0], y= data.iloc[:, 1], native_scale=1)
#ax.legend()
plt.show()
```

## Amplitudes

```python
data.columns
```

```python
downs = []
tops = []
amps = []
for col in data.columns[3:]:
    down = []
    top = []
    amp = []
    vals = np.array(data[col])
    for i in range(len(off)):
        top.append(np.mean(vals[off[i]-10:off[i]-2]))
        down.append(vals[on[i]-1])
        amp.append(top[i]-down[i])
    downs.append(down)
    tops.append(top)
    amps.append(amp)
    del down[13]
    del top[13]
    del amp[13]
downs = np.array(downs)
tops = np.array(tops)
amps = np.array(amps)
```
